### Task1: Merging 12 months of sales data into a single csv file

```python
# displaying them here
df = pd.read_csv('./Sales_Data/Sales_April_2019.csv')

# 2 dots .. goes back one folder, 1 dot . means current folder
files = [file for file in os.listdir('./Sales_Data')]

for file in files:
    print(file)
```

```python
# creating an empty dataframe
all_months_data = pd.DataFrame()
```

```python
# concatinate them together
for file in files:
    df = pd.read_csv("./Sales_Data/"+file, sep=',', error_bad_lines=False)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.head(5)
```

```python
all_months_data.to_csv('all_data_combined.csv', index=False)
all_months_data.head(20)
```

### Task2: Add a month column

Augment data with aditional columns

```python
# add another column specific for the month - this will make our analysis easier
# we want to get the first 2 numbers from the Order Date Column 
all_months_data['Month'] = all_months_data['Order Date'].str[0:2]
# we access the Order Date column by .str and then slicing the first 2 characters
all_months_data.head(15)
```

```python
# because we got a string we need to convert it to a numerical value 
all_months_data['Month'] = all_months_data['Month'].astype('int32')
all_months_data.head(15)
```

Cleaning up the data  (Drop rows of NaN)

error: cannot convert float NaN to integer

```python
# apparently we have some missing - NAN data that we have to clean up first.
# lets first find the NAN data first 
nan_df = all_months_data[all_months_data.isna().any(axis=1)]
nan_df.head(15)
```

```python
# the parameter how = 'all' will drop any row with that has all columns filled with NAN
# the parameter how = 'any' will drop any row with a single NAN in any column
all_months_data = all_months_data.dropna(how='all')
all_months_data.head(15)
```

```python
# invalid literal for int() with base 10: 'Or'
# we get a different error now so we need to figure out what we need to clean up to fix this issue 
all_months_data['Month'] = all_months_data['Month'].astype('int32')
all_months_data
```

Find Or and delete it 

error: invalid literal for int() with base 10: 'Or'

```python
# For whatever reason we get some rows with the same column names - they got duplicated throughout our Dataframe
temp_df = all_months_data[all_months_data['Order Date'].str[0:2] == 'Or']
temp_df.head()
```

```python
# done! we cleaned our df from Nan and Duplicate values
all_months_data = all_months_data[all_months_data['Order Date'].str[0:2] != 'Or']
all_months_data.head()
```

### Task3: Add a Sales Column

By adding the 'Month' column we can now filter by month, but we also need to be able to filter by Sales too!

To get the Sales we can multiply the 'Quantity Ordered' by the 'Price Each'

'Quantity Ordered' x 'Price Each' = per order Sales value

```python
all_months_data['Sales'] = all_months_data['Quantity Ordered'] * all_months_data['Price Each']
all_months_data.head(15)
```

looks like we have a little bit more cleaning to do!

error: can't multiply sequence by non-int of type 'str'

Lets convert columns to the correct type

```python
all_months_data.head(10)
```

```python
# have in mind that Quantity Ordered needs to be converted to int 
# where as Price Each needs to be converted to float!
# you can also use .astype but this way is better
all_months_data['Quantity Ordered'] = pd.to_numeric(all_months_data['Quantity Ordered'])
all_months_data.head(10)
```

```python
# Price Each needs to be converted to float!
all_months_data['Price Each'] = pd.to_numeric(all_months_data['Price Each'])
all_months_data.head(10)
```

```python
# lets now try again to make the creation of the new column with the product of the 2 columns
all_months_data['Sales'] = all_months_data['Quantity Ordered'] * all_months_data['Price Each']
all_months_data.head(10)
```

### Question 2: What city had the highest number of sales?

Task 5: Add a City Column

```python
# first step to solve this is to augment the data 
# we can add a new column - a City column
all_months_data.head()

# We have a Purchase Address column
# from that column we can see that in the middle of the values we have the City Name   

# We can see that the city is between 2 commas (,) 
# so what we can do is split on the commas and get the city 
```

```python
# let's use the .apply() method 
# this .apply() method allows us to run any function on our Dataframe

# so what we are doing here is we split the cell contencts of the address adress
# this gives us 3 values:  0 -> 917 1st St    1-> Dallas   2-> TX 75001
all_months_data['Column'] = all_months_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_months_data.head()
```

```python
# accidently named the column wrong so lets remove it 
all_months_data = all_months_data.drop(columns='Column')
# all_months_data.drop(columns='Column', inplace=True)
all_months_data.head()
```

```python
# not the most efficient way mostly
# alternative way
def get_city(address):
    return address.split(',')[1]

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x))

all_months_data.head()
```

The problem with just stating  a city 

the problem is that there are Cities with the same Name in different countries.

For example there are 2 Portlands in the in US.

Portland Maine and Portland Oregon

This means that we might end up with Duplicates but mean different areas.

So we should also grab the state

```python
def get_state(address):
    return address.split(',')[2]

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_months_data.head()
```

now lets remove the zip code, since its not needed

```python
# when you split on the white space you end up ultimately with the left side the results with an empty string

# we have white space ' ' before and after Texas so we get this:
# we get ['','TX','75001'] thats why we want the 2nd one, [1]

def city_state(address):
    return address.split(',')[2].split(' ')[1]


all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + city_state(x))

all_months_data.head()
```

```python
# lets also format this abit

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + city_state(x) + ')')

all_months_data.head()
```

f"{var}" strings  - string fromatting

```python
# another way to do this is with string formatting

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({city_state(x)})")

all_months_data.head()
```

```python
results = all_months_data.groupby('City').sum()
results
```
