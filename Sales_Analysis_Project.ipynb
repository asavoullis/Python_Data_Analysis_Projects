```python
# first step to solve this is to augment the data 
# we can add a new column - a City column
all_months_data.head()

# We have a Purchase Address column
# from that column we can see that in the middle of the values we have the City Name   

# We can see that the city is between 2 commas (,) 
# so what we can do is split on the commas and get the city 
```

```python
# let's use the .apply() method 
# this .apply() method allows us to run any function on our Dataframe

# so what we are doing here is we split the cell contencts of the address adress
# this gives us 3 values:  0 -> 917 1st St    1-> Dallas   2-> TX 75001
all_months_data['Column'] = all_months_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_months_data.head()
```

```python
# not the most efficient way mostly
# alternative way
def get_city(address):
    return address.split(',')[1]

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x))

all_months_data.head()
```

The problem with just stating  a city 

the problem is that there are Cities with the same Name in different countries.

For example there are 2 Portlands in the in US.

Portland Maine and Portland Oregon

This means that we might end up with Duplicates but mean different areas.

So we should also grab the state

```python
def get_state(address):
    return address.split(',')[2]

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_months_data.head()
```

```python
# lets also format this abit

all_months_data['City'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + city_state(x) + ')')

all_months_data.head()
```

### Question 5: What products are most often sold together? (Hard)

```python
all_months_data.head()
```

If you look at the dataset we have we can see that we have a column named 'Order ID'.

When we display the dataset we can see 2 orders with the same ID Number, 176560. 

This means that those 2 products, named on the 'Product' column, the Google Phone and Wired Headphones were sold in the same order together.

```python
# we can see here that the order id with number 176560 purchased 2 different products
all_months_data[all_months_data['Order ID'] == '176560' ]
```

The first thing we can do to approach this question is we need to know the spots in our dataframe that have duplicate order ids.

```python
# using df[column name].value_counts() we get a series with the values that appear most times
# in this case the order id 170873 purchased 5 different products
all_months_data['Order ID'].value_counts()
```

```python
# displaying the order id with number 160873
all_months_data[all_months_data['Order ID'] == '160873' ]
```

```python
# lets find only the duplicated data now
df_new = all_months_data[all_months_data['Order ID'].duplicated()]
df_new.head(10)
```

```python
# lets find only the duplicated data now
# lets keep the duplicated by using keep = False
df_new = all_months_data[all_months_data['Order ID'].duplicated(keep = False)]
df_new.head(10)
# https://stackoverflow.com/questions/43348194/pandas-select-rows-if-id-appear-several-time
```

Now lets group them together by creating a new column.

This new column will have the 'Products' that were bought in the same 'Order ID' in the same row.

Instead of having like many rows of each product we want 1 now.

```python
# transform works similarly to apply
df_new['Grouped']= df_new.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df_new.head()
# Referenced: https://stackoverflow.com/questions/27298178/concatenate-strings-from-several-rows-using-pandas-groupby
```

```python
# the problem here is that we have duplicated rows so lets drop the duplicated ones
df_new = df_new[['Order ID', 'Grouped']].drop_duplicates()
df_new
```

Now lets iterate over all the rows in the Grouped column and count those pairs and get that final count of what products are sold more often together.

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

# we need to count the pairs of what occurs together most frequently.
from itertools import combinations
from collections import Counter

# basically we make a list of comma separated values eg [[2,3], [2,3,4]] (list of lists)
# and then we update the count based on the combinations in the sublist

count = Counter()

for row in  df_new['Grouped']:
    row_list = row.split(',')
    # we want to use combinations of 2 hence (var, 2)
    count.update(Counter(combinations(row_list, 2)))

print(count)
```

As we can see the most common items sold together as a pair are: 

'iPhone' and 'Lightning Charging Cable'

```python
# lets make a nicer format of this
count.most_common(10)
```

```python
# or even more readable format:

for key, value in count.most_common(10):
    print(key, value)
```

As we can see here the 2 products that are most commonly bought together are :

'iPhone' and 'Lightning Charging Cable'

And they were bought 1005 times.

Followed by:

'Google Phone' and 'USB-C Charging Cable' that were bought 987 times togher

```python
# we could also check for a 3 pair, the 3 most common products sold together, because above it was for 2

count_3 = Counter()

for row in  df_new['Grouped']:
    row_list_3 = row.split(',')
    # we want to use combinations of 3 hence (var, 3)
    count_3.update(Counter(combinations(row_list_3, 3)))

for key, value in count_3.most_common(10):
    print(key, value)
```

We can see that when choosing a 3 pair combination the Google phone passes the iPhone.
