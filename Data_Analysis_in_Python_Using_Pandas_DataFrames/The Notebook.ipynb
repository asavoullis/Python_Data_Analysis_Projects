### Let's answer to some analytical questions 
    
Question 1 - In which year we have the `highest` and `lowest` number of movies?

Now finding out number of movies per year

```python
data.head()
```

```python
movie_count_per_year = data.groupby("title_year").size()
movie_count_per_year
```

```python
movie_count_per_year[2010]
```

```python
movie_count_per_year.max()
```

```python
movie_count_per_year.idxmax()
```

```python
movie_count_per_year.idxmin()
```

```python
movie_count_per_year.min()
```

Let's create a line chart to show the result

```python
ax = movie_count_per_year.plot(figsize = (8,6))
ax.set_xlabel("Years")
ax.set_ylabel("# of Movies")
```

Question 2 - In which year we have `minimum` and `maximum` average `imdb_score` ? 

Finding average imdb score per year 

```python
average_imdb_score_per_year = data.groupby("title_year")['imdb_score'].mean()
average_imdb_score_per_year
```

Let's Use a barchart to visualize the result

```python
ax = average_imdb_score_per_year.plot.bar(figsize = (18,8), title = "Average Imdb Score per Year")
ax.set_xlabel = "Year"
ax.set_ylabel = "Average Score per Year"
```
