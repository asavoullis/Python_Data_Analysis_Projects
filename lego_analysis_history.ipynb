### Question 1

we need to find what themes are star wars related

we can see that we have a parent_theme field in our first dataset (legosets [df]) and it matches the name field in our second dataset (parent_themes [themes])

so lets go ahead and JOIN them using JOIN or Merge

```python
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged.head(10)
# merge df and theme on the parent_theme and name columns
# we had a name column in both datasets so this ended up anding a prefix in both columns name_x and name_y
```

```python
# lets go ahead then and remove the name_y column since its not needed
merged = merged.drop(columns='name_y')
# alternative way:
# merged.drop(columns='name_y', inplace = True)   
merged
```

```python
# lets go ahead and filter the dataset to just the licensed dataframes
licensed = merged[merged['is_licensed'] == True]
# licensed is the dataset with all the True values in the is_licensed column
# licensed = from merged, what places in megrged the column is_licensed is True
licensed
```

```python
# we need to filter the null values of the column set number
# lets check how many are there
licensed[licensed['set_num'].isnull()].shape
# so as we can see we have 153 null or missing values in that column
```

```python
# we need to filter the null values of the set_num column 
# but we need to be carefull when dropping null values because other columns might have too
# for example the column num_parts. This column is not central to our analyses, so missing values are acceptable
# .dropna() drops all the null values from the dataset in any column
licensed = licensed.dropna(subset=['set_num'])
licensed
# which is correct because 1332 - 153 = 1179
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
# we make a new dataframe that has only the parent_theme set as Star Wars
star_wars
```

```python
# find the count of licensed and the count of star_wars 
star_wars.shape
# .shape: Returns a tuple of shape (Rows, columns) of dataframe/series
# .size: Returns the size of dataframe/series which is equivalent to total number of elements. That is rows x columns.
```

### Question 2

In which year was Star Wars not the most popular licensed theme (in terms of sets released that year)?

Save your answer as variable new_era in the form of an integer (e.g. 2012)

```python
# lets sort them by year and see how our dataset looks like
# we can use the licensed dataframe as we are looking for a licensed theme again
licensed_sorted = licensed.sort_values('year')
licensed_sorted
```

```python
licensed_sorted['Count'] = 1
# we make a new column called Count and set all the row values equal to 1 
licensed_sorted.head()
```

```python
# now lets group them now by year and by parent_theme - as we need a year (use group by)
licensed_sorted.groupby(['year', 'parent_theme']).head()
```

```python
# now lets group them now by year and by parent_theme - as we need a year (use group by)
summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum()
summed_df
# apparently we got a summation of all the True Values depending on the year and the parent theme in the is_licensed column
# because we got only True values it decided to count them as one and add them up
# We can verify this because we get the same result as the count column we made 
```

```python
summed_df = summed_df.reset_index()
summed_df
# now lets fix or add the index 
# this is something its good to do when you aggregate (combine or sum or caclulate or manipulate) the columns
# usually we use it after we use the groupby function
```

```python
# now lets get the max of 
max_df = summed_df.sort_values('Count', ascending=False).drop_duplicates(['year'])
max_df
```

```python
max_df.sort_values('year',inplace=True)
max_df
```

### Question 3

Which year had the most different parent_themes that were built
Which year had the most number of sets - the set_nums? - if we count the set numbers for each year
Break down number of sets by year

```python
cleaned_df = merged[~merged['set_num'].isnull()]
# before we used is null to find how many invalid records we had
# now we used
cleaned_df
```

```python
# lets add a new column to our new df and fill it with 1s
cleaned_df['count'] = 1
cleaned_df
```

```python
cleaned_df.groupby(['year']).sum().reset_index()
# now lets group by year,   sum it up,  reset index
```

```python
cleaned_df.groupby(['year']).sum().reset_index()[['year', 'count']]
# now lets take only the year and the count
# notice the double brackets at the end
```

```python
# lets also rename it
sets_per_year = cleaned_df.groupby(['year']).sum().reset_index()[['year', 'count']]
sets_per_year
```

```python
for index, row in sets_per_year.iterrows():
    print(row['year'], row['count'])
# we can see that there weren't that many type of sets through the 50s 
# we can also notice that in 1966 there was a sudden peak of 89, compared to 10 the previous year and 21 in the next year
# overall we can see that the there is an increasing trend as the years go on
```
